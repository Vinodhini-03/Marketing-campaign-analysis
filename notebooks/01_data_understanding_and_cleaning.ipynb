```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/marketing_data.csv")
data_dict = pd.read_csv("../data/marketing_data_dictionary.csv")
```

```python
df.head()
df.shape
df.info()
df.columns
```

```python
df.isnull().sum()
(df.isnull().mean() * 100).round(2)
```

```python
df_clean = df.copy()

df_clean["Income"] = df_clean["Income"].fillna(df_clean["Income"].median())
```

```python
df_clean.dropna(inplace=True)
```

```python

df_clean["Dt_Customer"] = pd.to_datetime(
    df_clean["Dt_Customer"],
    errors="coerce",
    dayfirst=True
)

```

```python
df_clean.dropna(subset=["Dt_Customer"], inplace=True)
df_clean["Dt_Customer"].dtype
```

```python
#age
df_clean["Age"] = 2024 - df_clean["Year_Birth"]
df_clean = df_clean[(df_clean["Age"] >= 18) & (df_clean["Age"] <= 100)]
```

```python
#customer tensure
df_clean["Customer_Tenure_Days"] = (
    df_clean["Dt_Customer"].max() - df_clean["Dt_Customer"]
).dt.days

```

```python
df_clean[["Dt_Customer", "Customer_Tenure_Days"]].head()
```

```python
df_clean["Customer_Since_Years"] = (
    df_clean["Customer_Tenure_Days"] / 365
).round(1)
```

```python
#total spend
df_clean["Total_Spend"] = (
    df_clean["MntWines"] +
    df_clean["MntFruits"] +
    df_clean["MntMeatProducts"] +
    df_clean["MntFishProducts"] +
    df_clean["MntSweetProducts"] +
    df_clean["MntGoldProds"]
)
```

```python
#total purchase
df_clean["Total_Purchases"] = (
    df_clean["NumWebPurchases"] +
    df_clean["NumCatalogPurchases"] +
    df_clean["NumStorePurchases"]
)
```

```python
df_clean["Children"] = df_clean["Kidhome"] + df_clean["Teenhome"]
```

```python
df_clean.shape
df_clean.info()
```

```python
output_path = "../data/marketing_data_cleaned.csv"
df_clean.to_csv(output_path, index=False)

output_path
```
