```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/marketing_data_cleaned.csv")
```

```python
df.shape
df.head()
```

```python
df["High_Income"] = np.where(df["Income"] > 75000, 1, 0)
df[['Income', 'High_Income']].head()
```

```python
df["Young_Customer"] = np.where(df["Age"] < 30, 1, 0)
df[['Age', 'Young_Customer']].head()
```

```python
df["Family_Customer"] = np.where(df["Children"] > 0, 1, 0)
df[['Children', 'Family_Customer']].head()
```

```python
df["Campaign_Responder"] = np.where(df["Response"] == 1, 1, 0)
df[['Response', 'Campaign_Responder']].head()
```

```python
df["High_Web_Engagement"] = np.where(
    df["NumWebVisitsMonth"] > 5, 1, 0
)
df[['NumWebVisitsMonth', 'High_Web_Engagement']].head()
```

```python
spend_threshold = df["Total_Spend"].quantile(0.90)

df["High_Spender"] = np.where(
    df["Total_Spend"] > spend_threshold, 1, 0
)
df[['Total_Spend', 'High_Spender']].head()
```

```python
df["Age_Band"] = pd.cut(
    df["Age"],
    bins=[18, 30, 45, 60, 100],
    labels=["18-29", "30-44", "45-59", "60+"]
)
df[['Age', 'Age_Band']].head()
```

```python
df["Income_Band"] = pd.cut(
    df["Income"],
    bins=[0, 30000, 60000, 90000, np.inf],
    labels=["Low", "Medium", "High", "Very High"]
)
df[['Income', 'Income_Band']].head()
```

```python
df[
    [
        "High_Income",
        "Young_Customer",
        "Family_Customer",
        "Campaign_Responder",
        "High_Web_Engagement",
        "High_Spender"
    ]
].mean().round(2)
```

```python
output_path = "../data/marketing_data_final.csv"
df.to_csv(output_path, index=False)

output_path
```
