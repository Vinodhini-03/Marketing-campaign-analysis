```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

```

```python
df = pd.read_csv("../data/marketing_data_cleaned.csv")

```

```python
df.shape, df.columns.tolist()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Income"], bins=30, kde=True)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(y="Education", data=df)
plt.title("Customer Distribution by Education Level")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(y="Marital_Status", data=df)
plt.title("Customer Distribution by Marital Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Income", y="Total_Spend", data=df)
plt.title("Income vs Total Spend")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Total_Spend", data=df)
plt.title("Age vs Total Spend")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Education", y="Total_Spend", data=df)
plt.title("Total Spend by Education Level")
plt.show()
```

```python
channel_cols = [
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases"
]

df[channel_cols].mean().plot(kind="bar", figsize=(6,4))
plt.title("Average Purchases by Channel")
plt.ylabel("Average Number of Purchases")
plt.show()
```

```python
sns.countplot(x="Response", data=df)
plt.title("Campaign Response Distribution")
plt.show()
```

```python
sns.boxplot(x="Response", y="Income", data=df)
plt.title("Income vs Campaign Response")
plt.show()
```

```python
sns.boxplot(x="Response", y="Total_Spend", data=df)
plt.title("Total Spend vs Campaign Response")
plt.show()
```
